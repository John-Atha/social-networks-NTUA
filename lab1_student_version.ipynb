## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)

# REGural graph => { degree(v) = d for each node v of the graph }
n = 41
d = 4
p = 0 # not random network
REG = nx.watts_strogatz_graph(n, d, p) 

nx.draw_circular(REG)
```

```python
# many edges are overlapping
# I use a different draw method so that the degree of each node is more easily observed
nx.draw_random(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)

# RandomGraph(Erdos-Renyi) => { pick a random graph from the set of all graphs with n nodes and M edges }
n=41 
M = 150
RGER = nx.gnm_random_graph(n, M)

nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
# RandomGeometricGraph => { nodes are positioned randomly, edges are created between nodes with distance<R }
n=41
R=0.35

RGG = nx.random_geometric_graph(n, radius=R)

# following code for visualization from:
# https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

pos = nx.get_node_attributes(RGG, "pos")
# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)

# ScaleFree => { degree distribution follows a power-law => P[deg(u)==k]=k^(-γ) }

n = 41
d = 4
SF = nx.barabasi_albert_graph(n, d)

nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)

# SmallWorld => { at first  create a regular graph and then randomly rewire each edge with p=0.25 } 

n = 41
d = 4
p = 0.25
SW = nx.watts_strogatz_graph(n, d, p) 

nx.draw_circular(SW)
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):


```python
graphs = [
    {
        'graph': REG,
        'name': 'REG',
    },
    {
        'graph': RGER,
        'name': 'RGER',
    },
    {
        'graph': RGG,
        'name': 'RGG',
    },
    {
        'graph': SF,
        'name': 'SF',
    },
    {
        'graph': SW,
        'name': 'SW',
    },
]
```

#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.


```python
def cc(G):
    '''
    * for the graph G, return a dict with the clustering coefficient of each node
    '''
    d = { key: round(val, 3) for key, val in nx.clustering(G).items() } 
    return d

plt.figure(1)
plt.title("Clustering Coefficient Distribution")
plt.xlabel('Clustering Coefficient')
plt.ylabel('Number of Nodes')

for g in graphs:
    d = cc(g['graph'])
    print(f"--> {g['name']}:")
    print(d)
    plt.hist(d.values(), bins=10)
plt.legend([g['name'] for g in graphs])
    
```

#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.


```python
def cc_graph(G):
    '''
    * for the graph G, return the mean value and the variance of the clustering coefficients of its nodes
    '''
    d = cc(G)
    mean = round(
        sum(d.values())/len(d),
        3
    )
    variance = round(
        sum([(x/mean)**2 for x in d.values()])/len(G.nodes()),
        3
    )
    return mean, variance

print("Graph\t| mean\t| variance")
print('--------------------------')
for g in graphs:
    mean, variance = cc_graph(g['graph'])
    print(f"{g['name']}:\t| {mean}\t| {variance}")
    
    
```

#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 


```python
def min_max_cc(G):
    '''
    * for the graph G, return the pairs (node, cc_value) for the nodes with the minimum and the maximum clustering coefficient values
    '''
    d = cc(g['graph'])
    minim = min(d.items(), key=lambda x: x[1])
    maxim = max(d.items(), key=lambda x: x[1])
    return minim, maxim

print("graph\t| min_cc (node, cc) \t| max_cc (node, cc)")
print('---------------------------------------------------')
for g in graphs:
    minim, maxim = min_max_cc(g['graph'])
    print(f"{g['name']}\t| {minim}\t\t| {maxim}")
```

#### Να σχολιαστούν τα αποτελέσματα.

* Ιδιαίτερο ενδιαφέρον παρουσιάζει η γραφική παράσταση όπου φαίνονται οι κατανομές των τιμών των clustering coefficients για τους κόμβους όλων των γράφων:
    * Όπως είναι αναμενόμενο, στον `κανονικό γράφο`, ο clustering coefficient είναι σταθερός για όλους τους κόμβους.
    * Στον `random geometric graph`, οι τιμές του clustering coefficient είναι σημαντικά μεγάλες (πάνω από 0.5 σε 
      όλους τους κόμβους), κάτι που διαισθητικά βγάζει νόημα, αφού οι κόμβοι σχηματίζουν γειτονιές με βάση τις 
      γεωμετρικές αποστάσεις τους.
    * Στον γράφο `scale-free`, όλοι οι κόμβοι έχουν σχετικά μικρό συντελεστή, αφού η πιθανότητα ένας κόμβος να έχει 
      μεγάλο βαθμό (και άρα πολλούς γείτονες, και άρα μεγάλο συντελεστή) είναι εκθετικά μικρή.

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):

```python
graphs = [
    {
        'graph': REG,
        'name': 'REG',
    },
    {
        'graph': RGER,
        'name': 'RGER',
    },
    {
        'graph': RGG,
        'name': 'RGG',
    },
    {
        'graph': SF,
        'name': 'SF',
    },
    {
        'graph': SW,
        'name': 'SW',
    },
]
```

#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.

```python
def compute_path_lengths(G):
    '''
    * uses the `nx.all_pairs_shortest_path_length` to returns the shortest path lengths for all the pairs
    '''
    return dict(nx.all_pairs_shortest_path_length(G))

def extract_mean_var(G, d):
    '''
    * extracts the `mean` and `variance` of all the shortest paths lengths
    '''
    weights = [d[u][v] for u in G.nodes() for v in G.nodes() if u!=v]
    mean = round(
        sum(weights)/len(weights),
        3
    )
    variance = round(
        sum([(x-mean)**2 for x in weights])/len(weights),
        3
    )
    return mean, variance


print("Graph\t| mean\t| variance")
print('---------------------------')
for g in graphs:
    path_lengths = compute_path_lengths(g['graph'])
    mean, variance = extract_mean_var(g['graph'], path_lengths)
    print(f"{g['name']}\t| {mean}\t| {variance}")
        
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    '''
    * for each node
        * counts the number of other nodes that are connected to it with min_path_length = diameter
    * returns the nodes with the max counter and the connecting nodes that lead to this score
    '''
    path_lengths = compute_path_lengths(G)
    diameter = nx.algorithms.distance_measures.diameter(G)
    score = dict()
    
    # fill the score dict for each node
    for u in G.nodes():
        for v in G.nodes():
            if u<v:
                if path_lengths[u][v] == diameter:
                    score[u] = score[u]+1 if score.get(u) else 1
                    score[v] = score[v]+1 if score.get(v) else 1

    # find the max score and the nodes accomplishing it
    max_score = max(score.values())
    max_nodes = [node for node in score.keys() if score[node]==max_score]
    
    # add the nodes `u` with min_path_length(v, u)=diameter to the directory of the best nodes `v`
    nodes = {u: set([v for v in path_lengths[u] if path_lengths[u][v]==diameter]) for u in max_nodes}
    return nodes               
        
for g in graphs:
    nodes = find_nodes(g['graph'])
    print(f"{g['name']}:")
    for item in sorted(list(nodes.items())):
        print(f"  --> node {item[0]} (with {len(item[1])} nodes):", item[1])
    
    
```

#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 

```python
def find_periphery_from_scratch(G):
    '''
    * computes the periphery without using the built-in method of the networkx library
    '''
    path_lengths = compute_path_lengths(G)
    diameter = nx.algorithms.distance_measures.diameter(G)
    periphery = set()
    for u in G.nodes():
        eccentricity = max(path_lengths[u].values())
        if eccentricity==diameter:
            periphery.add(u)
    return periphery

def find_periphery(G):
    '''
    * computes the periphery using the built-in method of the library
    '''
    return nx.periphery(G)

# I store the results in the `peripheries` dictionary
# I will use it later to compare the graphs
peripheries = dict()

for g in graphs:
    periphery_built_in = set(find_periphery(g['graph']))
    periphery_from_scratch = set(find_periphery_from_scratch(g['graph']))
    print(f"{g['name']}:")
    print(f"  periphery: {periphery_from_scratch}")
    print('--------------------------------------------')
    assert periphery_built_in == periphery_from_scratch
    peripheries[g['name']] = periphery_from_scratch
```

#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  

```python
def find_center_from_scratch(G):
    '''
    * computes the center without using the built-in method of the networkx library
    '''
    path_lengths = compute_path_lengths(G)
    radius = nx.radius(G)
    center = set()
    for u in G.nodes():
        eccentricity = max(path_lengths[u].values())
        if eccentricity==radius:
            center.add(u)
    return center

def find_center(G):
    '''
    * computes the center using the built-in method of the library
    '''
    return nx.center(G)

# I store the results in the `centers` dictionary
# I will use it later to compare the graphs
centers = dict()

for g in graphs:
    center_built_in = set(find_center(g['graph']))
    center_from_scratch = set(find_center_from_scratch(g['graph']))
    print(f"{g['name']}:")
    print(f"  center: {center_from_scratch}")
    print('--------------------------------------------')
    assert center_built_in == center_from_scratch
    centers[g['name']] = center_from_scratch
```

* Παρατηρώ ότι στον Regural γράφο όλοι οι κόμβοι ανήκουν και στην περιφέρεια και στο κέντρο του.
    * Αυτό είναι αναμενόμενο λόγω της συμμετρίας που τον χαρακτηρίζει.
* Ο τυχαίος Erdos-Renyi γράφος, παρατηρώ ότι έχει μικρή περιφέρεια και μεγάλο κέντρο.
    * Δηλαδή διαισθητικά λίγα μέγιστα μονοπάτια φτάνουν το μήκος της διαμέτρου, με τα περισσότερα να είναι κοντά 
      στην ακτίνα.
* Παρατηρώ ότι ο τυχαίος γεωμετρικός γράφος έχει σχετικά χαμηλές και τις δύο μετρικές.
    * Άρα τα περισσότερα μέγιστα μονοπάτια του είναι ενδιάμεσου μήκους (ενδιάμεσα στην διάμετρο και στην ακτίνα).

    * Αυτό διαισθητικά αιτιολογείται από το γεγονός ότι οι ακμές-"συντομεύσεις" οδηγούν σε ένα "μοίρασμα" των κόμβων 
      του γράφου σην περιφέρεια και στο κέντρο του.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.

```python
def find_degree(G):
    return G.degree()

plt.figure(1)
plt.title('Degrees distribution')
plt.xlabel('degree')
plt.ylabel('# nodes')

degrees = dict()
# one plot combining all the distributions
for g in graphs:
    graph, name = g.values()
    degrees[name] = [deg[1] for deg in find_degree(graph)]
    plt.hist(degrees[name], bins=10)
    
plt.legend([g['name'] for g in graphs])

# one seperate plot for each distribution to avoid previous visual overlappings
c = 2
for g in graphs:
    graph, name = g.values()
    plt.figure(c)
    plt.xlabel('degree')
    plt.ylabel('# nodes')
    plt.hist(degrees[name], bins=10)
    plt.legend([name])
    c += 1
    
```

##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.

```python
print("graph\t| avg degree\t| variance of degrees")
print('---------------------------------------------')
for g in graphs:
    graph, name = g.values()
    g_degrees = [deg[1] for deg in find_degree(graph)]
    avg = round(sum(g_degrees)/len(g_degrees), 2)
    var = round(sum([(val-avg)**2/len(g_degrees) for val in g_degrees]), 2)
    print(f"{name}\t| {avg}\t\t| {var}")
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.

```python
def find_katz(G):
    '''
    * return a dictionary with the katz centrality for each node of the graph G
    '''
    # compute alpha
    lambda_max = nx.adjacency_spectrum(G).max().real
    alpha = round(1/lambda_max - 0.01, 3)
    centr = nx.katz_centrality(G, alpha, beta=1.0)
    return centr

katz = dict()
plt.figure(1)
plt.title("Katz centralities distributions")
plt.xlabel("Katz centrality")
plt.ylabel("# nodes")
# generate a plot compbining the katz centrality distributions of all the graphs
for g in graphs:
    graph, name = g.values()
    katz[name] = find_katz(graph).values()
    plt.hist(katz[name], bins=10)

plt.legend([g['name'] for g in graphs]) 

# generate a separate plot for each graph
c = 2
for g in graphs:
    graph, name = g.values()
    plt.figure(c)
    plt.title("Katz centralities distribution")
    plt.xlabel("Katz centrality")
    plt.ylabel("# nodes")
    plt.hist(katz[name], bins=10)
    plt.legend([name])
    c += 1
```

##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.

```python
print("graph\t| mean\t| variance")
print("---------------------------")

for g in graphs:
    graph, name = g.values()
    katz_g = find_katz(graph).values()
    avg = round(sum(katz_g) / len(katz_g), 6)
    var = round(sum([(x-avg)**2/len(katz_g) for x in katz_g]), 6)
    print(f"{name}\t| {avg}\t| {var}")
    
    
```

```python
def find_most_central(G, centralities):
    max_degree = 0
    max_degree_nodes = []
    for node_centr in centralities:
        node, degree = node_centr
        if max_degree == degree:
            max_degree_nodes.append(node)
        elif max_degree < degree:
            max_degree = degree
            max_degree_nodes = [node]
    return max_degree_nodes

def find_central(G):
    degree_centralities = nx.degree_centrality(G).items()
    closeness_centralities = find_closeness(G).items()
    betweenness_centralities = find_betweenness(G).items()
    katz_centralities = find_katz(G).items()
    print(f"Most center nodes of {name}:")
    print("-->degree centrality:", find_most_central(G, degree_centralities))
    print("-->closeness centrality:", find_most_central(G, closeness_centralities))
    print("-->betweenness centrality:", find_most_central(G, betweenness_centralities))
    print("-->katz centrality:", find_most_central(G, katz_centralities))
    print("----------------------------")
for g in graphs:
    graph, name = g.values()
    find_central(graph)
        
            
        
```
